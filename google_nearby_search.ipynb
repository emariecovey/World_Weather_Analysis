```python
# Dependencies and Setup
import requests
import gmaps

# Import API key
from config import g_key
```

```python
# Set the parameters to search for a hotel in Paris.
params = {
    "radius": 5000,
    "types": "lodging",
    "key": g_key,
    "location": "48.8566, 2.3522"}

base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"

#make a request:
hotels = requests.get(base_url, params=params).json()
hotels
```

```python
hotels.keys() #keys to json dictionary are ['html_attributions', 'next_page_token', 'results', 'status']

len(hotels["results"]) #shows we have 20 results

hotels["results"][0]["name"] #gets name of first hotel

hotels["results"][4]["opening_hours"].values() #says if fifth hotel was currently open (t/f)

hotels["results"][0].keys() #finding all the different atributes of the results section
```
